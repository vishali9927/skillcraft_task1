```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Simulated dataset
data = {
    'Age Group': ['0 - 20 Years', '21 - 44 Years', '45+ Years'],
    'Population (millions)': [512, 607, 198]
}

df = pd.DataFrame(data)
```

```python
# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Age Group', y='Population (millions)', data=df, palette='viridis')

plt.title("India's Population Distribution by Age (2022)", fontsize=16)
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population (in millions)", fontsize=12)
plt.tight_layout()
plt.show()
```
